```python
import pandas as pd 
import numpy as np 
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv(r"Credit card transactions - India - Simple.csv")
print(data)
```

```python
data.head()
```

```python
total_amount_by_gender = data.groupby('Gender')['Amount'].sum()
print("Total amount spent by gender:")
print(total_amount_by_gender)
```

```python
avg_amount_by_card_type = data.groupby('Card Type')['Amount'].mean()
print("\nAverage transaction amount by card type:")
print(avg_amount_by_card_type)
```

```python
total_amount_by_city = data.groupby('City')['Amount'].sum().sort_values(ascending=False)
print("\nTotal amount spent by city:")
print(total_amount_by_city)
```

```python
total_amount_by_expense_type = data.groupby('Exp Type')['Amount'].sum().sort_values(ascending=False)
print("\nTotal amount spent by expense type:")
print(total_amount_by_expense_type)
```

```python
# Visualizations
plt.figure(figsize=(12, 6))

# Total amount spent by gender
plt.subplot(2, 2, 1)
sns.barplot(x=total_amount_by_gender.index, y=total_amount_by_gender.values)
plt.title('Total amount spent by gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount')

# Average transaction amount by card type
plt.subplot(2, 2, 2)
sns.barplot(x=avg_amount_by_card_type.index, y=avg_amount_by_card_type.values)
plt.title('Average transaction amount by card type')
plt.xlabel('Card Type')
plt.ylabel('Average Amount')

# Total amount spent by city
plt.subplot(2, 2, 3)
sns.barplot(x=total_amount_by_city.head(10).index, y=total_amount_by_city.head(10).values)
plt.title('Total amount spent by city (Top 10)')
plt.xlabel('City')
plt.ylabel('Total Amount')
plt.xticks(rotation=90)

# Total amount spent by expense type
plt.subplot(2, 2, 4)
sns.barplot(x=total_amount_by_expense_type.head(10).index, y=total_amount_by_expense_type.head(10).values)
plt.title('Total amount spent by expense type (Top 10)')
plt.xlabel('Expense Type')
plt.ylabel('Total Amount')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
